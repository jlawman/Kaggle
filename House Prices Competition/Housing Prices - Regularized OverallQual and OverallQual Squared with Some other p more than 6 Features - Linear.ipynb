```python
#Highly influential select features with (p>.6)
hi_inf_train=train[['GrLivArea', 'GarageArea', 'TotalBsmtSF', '1stFlrSF','OverallQual']]
hi_inf_train.head()
```

```python
hi_inf_train['OverallQualReg']=hi_inf_train['OverallQual']*300
hi_inf_train['OverallQualSquared']=(hi_inf_train['OverallQualReg']*hi_inf_train['OverallQualReg'])/5000 #squared and divided by 5,000 to regularize
```

```python
hi_inf_train=hi_inf_train.drop('OverallQual',axis=1)
hi_inf_train.head()
```

```python
hi_inf_train[hi_inf_train.isnull().any(axis=1)] #Checking for null entries
```

```python
#hi_inf_train=hi_inf_train.fillna(0)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()

x_train=hi_inf_train
y_train=train['SalePrice']


reg.fit(x_train,y_train)
reg.score(x_train,y_train)
```

```python
reg.coef_
```

```python
hi_inf_test=test[['GrLivArea', 'GarageArea', 'TotalBsmtSF', '1stFlrSF','OverallQual']]

#regularize the qual
hi_inf_test['OverallQualReg']=hi_inf_test['OverallQual']*300
hi_inf_test['OverallQualSquared']=(hi_inf_test['OverallQualReg']*hi_inf_test['OverallQualReg'])/5000 #squared and divided by 5,000 to regularize
```

```python
hi_inf_test=hi_inf_test.drop('OverallQual',axis=1)
x_test=hi_inf_test
```

```python
x_test[x_test.isnull().any(axis=1)]  #Look for null values
```

```python
x_test=x_test.fillna(0) #Fill Null garage area and basement with 0
x_test[x_test.isnull().any(axis=1)] 
```

```python
y_pred=reg.predict(x_test) #Abs is a temporary fix to deal with the prices which, for some reason, were negative and so invalid.
```

```python
my_submission=pd.DataFrame(y_pred)
my_submission['Id']=test['Id']
my_submission.columns=['SalePrice','Id']
my_submission=my_submission[['Id','SalePrice']]
```
